```python
# checking outliers on numeric columns:

df.describe()
```

```python
numeric_col = ['symboling','wheel-base','length','width','height','curb-weight','engine-size','compression-ratio',
               'city-mpg','highway-mpg']
```

```python
for i in numeric_col:
    sns.boxplot(df[i])
    plt.show()
```

```python
#treating outliers:
```

```python
def plot(x):
    sns.boxplot(df[x])
    plt.show()
    return
```

```python
for x in numeric_col:
    
    q3,q1 = np.percentile(df[x],[75,25])
    q4 = np.percentile(df[x],[99])
    df.loc[df[x]>q4[0],x]=q4[0]
    plot(x)
```
