# step:2 - checking null values and treating:

```python
data.isnull().sum()/len(data)*100
```

```python
## checking column type 2:

data['type2'].describe() 
```

```python
# treating null values:   #object are treated by mode and (int/float) with mean and median to fill null values. 

data.fillna(data['type2'].mode()[0], inplace = True)
```

```python
data['type2'].describe() 
```

```python
data.info()
```

```python
data.describe()
```

```python
# checking outliers on numeric columns:

numeric_col = ['attack','defense','sp_attack','sp_defense','speed','generation','base_total','hp','is_legendary']

for i in numeric_col:
    sns.boxplot(data[i])
    plt.show()
```

```python
# treating outliers:

def plot(x):
    sns.boxplot(data[x])
    plt.show()
    return
```

```python
for x in numeric_col:
    
    q3,q1 = np.percentile(data[x],[75,25])
    
    q4 = np.percentile(data[x],[99])
    
    data.loc[data[x]>q4[0],x]=q4[0]
    
    plot(x)
```
